# 12. Classify the employees into 3 categories.Category 
1 : employees below 18.5 bmi - underweightCategory 
2 : employees with bmi in the range of 18.5 to 24.9 - fitCategory 
3 : employee with bmi over 24.9 - overweight 
What is the average expenditure of employees in each category

```python
df_underweightCategory = df[df['bmi'] < 18.5]
df_underweightCategory_expenditure = df_underweightCategory['expenditure'].mean()
df_fitCategory  = df[  ( df['bmi'] > 18.5) & (df['bmi'] < 24.9)      ] 
df_fitCategory_expenditure = df_fitCategory['expenditure'].mean()
df_overweight = df[df ['bmi'] > 24.9]
df_overweight_expenditure = df_overweight['expenditure'].mean()

data = [['data_underweight',df_underweightCategory_expenditure],
       ['fit',df_fitCategory_expenditure],
        ['overweight',df_overweight_expenditure]]

table = pd.DataFrame(data,columns=['bmi_catagory','avg_expenditure'])
print (table)

sns.barplot(x='bmi_catagory',y='avg_expenditure',data= table)
```

# 16. Draw pie chart for alcohol consumption habits

```python
df16 = pd.pivot_table(data = df ,index = 'alcohol', values= 'expenditure',
                      aggfunc=np.sum )
df16 = df16.reset_index()
df16
```

```python
alcohol = []
for i in df16['alcohol']:
    alcohol.append(i)
alcohol
```

```python
expenditure = []
for i in df16['expenditure']:
    expenditure.append(i)
expenditure
```

```python
plt.pie(x=expenditure ,labels=alcohol ,autopct='%1.1f%%')
plt.show()
```
